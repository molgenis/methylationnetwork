```python
import pandas as pd
from sklearn.metrics import plot_roc_curve
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
model1_test_path = "./supplementary_materials/data/eqtm_predictions/model1_test_data.tsv"
model2_test_path = "./supplementary_materials/data/eqtm_predictions/model2_test_data.tsv"

model1_path = "./supplementary_materials/models/eqtm_predictions/model1.pkl"
model2_path = "./supplementary_materials/models/eqtm_predictions/model2.pkl"
```

```python
model1 = pickle.load(open(model1_path, 'rb'))
model1_test_data = pd.read_csv(model1_test_path, sep='\t', index_col=0)
model1_features = [item for item in model1_test_data if item not in ['model1_prob', 'label']]
plot_roc_curve(model1, model1_test_data[model1_features], model1_test_data['label'])
```

```python
model2 = pickle.load(open(model2_path, 'rb'))
model2_test_data = pd.read_csv(model2_test_path, sep='\t', index_col=0)
model2_features = [item for item in model2_test_data if item not in ['model2_prob', 'label']]
plot_roc_curve(model2, model2_test_data[model2_features], model2_test_data['label'])
```
